#데이터 준비

```python
import pandas as pd
wine_red = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/winequality-red.csv', sep=';')
wine_white = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/winequality-white.csv', sep=';')
```

```python
wine_red['color'] = 1
wine_white['color'] = 0 #레드, 화이트 와인을 구별하기 위해 column을 추가
```

```python
wine = pd.concat([wine_red, wine_white])
wine.reset_index(drop=True, inplace=True)
wine.info()
```

```python
wine.head()
```

```python
wine.info()
```

```python
wine.describe()
```

```python
data = wine[['fixed acidity',	'volatile acidity',	'citric acid',	'residual sugar',	'chlorides',	'free sulfur dioxide',	'total sulfur dioxide',	'density',	'pH',	'sulphates',	'alcohol', 'quality']].to_numpy()
target = wine['color'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
#test_size = 0.2 ==> 샘플의 20%정도를 테스트 세트로 나눔
train_input, test_input, train_target, test_target = train_test_split(data, target, stratify = target, test_size=0.2, random_state = 42)
```

```python
print(train_input.shape, test_input.shape)
```

```python
LABELS = ["Red_Wine", "White_Wine"]

import matplotlib.pyplot as plt

count_classes = pd.value_counts(wine['color'], sort = True)
count_classes.plot(kind='bar', rot = 0)
plt.title("Wine distribution")
plt.xticks(range(2), LABELS)
plt.xlabel("Wine class")
plt.ylabel("Frequency")
```

```python
from sklearn.preprocessing import StandardScaler
#표준화
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

#결정트리

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(train_scaled, train_target)

print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(10,7))
plot_tree(dt)
plt.show() #맨위는 루트노드, 아래는 리프노드
```

```python
plt.figure(figsize=(30,17))
plot_tree(dt, max_depth=3, filled=True, feature_names=['fixed acidity',	'volatile acidity',	'citric acid',	'residual sugar',	'chlorides',	'free sulfur dioxide',	'total sulfur dioxide',	'density',	'pH',	'sulphates',	'alcohol', 'quality'])
plt.show() #value -> [(음성 개수), (양성 개수)] // 양성이 커질 수록 파랑, 음성이 커질수록 주황
```

#랜덤 포레스트

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier
import numpy as np

rf = RandomForestClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(rf, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
rf.fit(train_input, train_target)
print(rf.feature_importances_)
```

```python
rf = RandomForestClassifier(oob_score=True, n_jobs=-1, random_state=42)
#oob_score : 부트스트랩 샘플에 포함되지 않는 샘플
rf.fit(train_input, train_target)
print(rf.oob_score_)
```

#엑스트라 트리

```python
from sklearn.ensemble import ExtraTreesClassifier

et = ExtraTreesClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(et, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
et.fit(train_input, train_target)
print(et.feature_importances_)
```

#그래디언트 부스팅


```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
gb = GradientBoostingClassifier(n_estimators=500, learning_rate=0.2, random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
gb.fit(train_input, train_target)
print(gb.feature_importances_)
```

#히스토그램 기반 그래디언트 부스팅 모델을 만들고 최적화

```python
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.experimental import enable_hist_gradient_boosting
hgb = HistGradientBoostingClassifier(random_state=42)
scores = cross_validate(hgb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
from sklearn.inspection import permutation_importance

hgb.fit(train_input, train_target)
result = permutation_importance(hgb, train_input, train_target, n_repeats=10,
                                random_state=42, n_jobs=-1)
print(result.importances_mean)
```

```python
result = permutation_importance(hgb, test_input, test_target, n_repeats=10,
                                random_state=42, n_jobs=-1)
print(result.importances_mean)
```

```python
hgb.score(test_input, test_target)
```
